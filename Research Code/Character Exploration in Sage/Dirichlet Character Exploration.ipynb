```python
n = 1000
Zn = Zmod(n)
G = DirichletGroup(n)
```

```python
def interested_values(group, numm):
    lst = []
    for g in group:
        if ZZ(g) > sqrt(numm):
            break
        lst.append(g)
    return [g for g in lst if gcd(ZZ(g), numm) == 1]

def short_character_sum(char, relevant_set):
    values = [char(relevant_set[0])]
    for i in range(1, len(relevant_set)):
        values.append(char(relevant_set[i])+values[i-1])
    return [CC(v) for v in values]

def find_character_that_returns_to_0(char_group, relevant_set):
    interesting_chars = []
    for char in char_group:
        values = short_character_sum(char, relevant_set)
        for val in values:
            if abs(val) == 0:
                interesting_chars.append(char)
                break
    return interesting_chars

def number_of_0_values(char, relevant_set):
    i = 0
    values = short_character_sum(char, relevant_set)
    for v in values:
        if v == 0:
            i += 1
    return i
```

```python
hold = interested_values(Zn, n)
```

```python
collection_data = []
```

```python
for n in range(100, 1001):
    Zn = Zmod(n)
    G = DirichletGroup(n)
    relevant_things = interested_values(Zn, n)
    data = find_character_that_returns_to_0(G, relevant_things)
    collection_data.append((n, len(data)/G.order(), sum([number_of_0_values(g, relevant_things) for g in data])/G.order()))
    print("n =", n, ", fraction of chars that have 0 =", collection_data[-1][1], ", average number of zeros =", collection_data[-1][2])
```
